```python
# Create a copy and calculate half-hourly consumption
df_copy = df.copy()

# Calculate gross_consumption_half_hourly: difference between current and 2 rows back (30 min interval)
# Only apply when both current and 2-rows-back values are not null
df_copy["gross_consumption_half_hourly"] = np.where(
    df_copy["kWh del (kWh)"].notna() & df_copy["kWh del (kWh)"].shift(2).notna(),
    df_copy["kWh del (kWh)"] - df_copy["kWh del (kWh)"].shift(2),
    np.nan
)

# Display the result
print(df_copy[["kWh del (kWh)", "gross_consumption_half_hourly"]].head(10))
```
