## Imported Libraries

```python
import pandas as pd
```

## Loading Dataset

```python
# Load CSV file
df = pd.read_csv("customer_data.csv")
```

## Processing Data

```python
# Convert timestamp column (adjust name if needed)
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

```

```python
# Create complete half-hour timeline
full_range = pd.date_range(
    start='2024-01-01 00:00',
    end='2024-12-31 23:30',
    freq='30min'
)

full_df = pd.DataFrame({'Timestamp': full_range})

# Merge with full timeline
merged = full_df.merge(df, on='Timestamp', how='left')

# Detect missing rows
missing = merged[merged['GROSS Consumption (kWh)'].isna()]

print("Missing rows:")
print(missing)

# Optional interpolation for missing values
merged['GROSS Consumption (kWh)'] = (
    merged['GROSS Consumption (kWh)']
    .interpolate()
)

# Save cleaned dataset back to CSV
merged.to_csv("cleaned_data.csv", index=False)
```
